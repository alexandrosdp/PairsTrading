```python
import pandas as pd
import numpy as np
import os
```

```python
results_BTC_WBTC = pd.read_csv('results/summary_measures BTC_USDT_2024_1m and WBTC_USDT_2024_1m.csv')
results_ETH_WBETH = pd.read_csv('results/summary_measures WBETH_USDT_2024_1m and ETH_USDT_2024_1m.csv')
results_Gala_GRT = pd.read_csv('results/summary_measures GALA_USDT_2024_1m and GRT_USDT_2024_1m.csv')
results_XRP_ADA = pd.read_csv('results/summary_measures XRP_USDT_2024_1m and ADA_USDT_2024_1m.csv')
```

```python
results_XRP_ADA
```

```python
#Concatenate all dataframes
results = pd.concat([results_BTC_WBTC, results_ETH_WBETH, results_Gala_GRT, results_XRP_ADA], axis=0)
results.reset_index(drop=True, inplace=True)
```
