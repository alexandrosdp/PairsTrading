```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pre Processing
---

```python
btc_usdt = pd.read_csv('order_book_data/BTCUSDT_1s.csv')
wbtc_usdt = pd.read_csv('order_book_data/WBTCUSDT_1s.csv')
eth_usdt = pd.read_csv('order_book_data/ETHUSDT_1s.csv')
wbeth_usdt = pd.read_csv('order_book_data/WBETHUSDT_1s.csv')
```

```python
#Order by local_timestamp
btc_usdt = btc_usdt.sort_values(by='local_timestamp')
wbtc_usdt = wbtc_usdt.sort_values(by='local_timestamp')
eth_usdt = eth_usdt.sort_values(by='local_timestamp')
wbeth_usdt = wbeth_usdt.sort_values(by='local_timestamp')


#Drop Unamed column
btc_usdt = btc_usdt.drop(columns=['Unnamed: 0'])
wbtc_usdt = wbtc_usdt.drop(columns=['Unnamed: 0'])
eth_usdt = eth_usdt.drop(columns=['Unnamed: 0'])
wbeth_usdt = wbeth_usdt.drop(columns=['Unnamed: 0'])

#Convert local_timestamp to datetime
btc_usdt['local_timestamp'] = pd.to_datetime(btc_usdt['local_timestamp'])
wbtc_usdt['local_timestamp'] = pd.to_datetime(wbtc_usdt['local_timestamp'])
eth_usdt['local_timestamp'] = pd.to_datetime(eth_usdt['local_timestamp'])
wbeth_usdt['local_timestamp'] = pd.to_datetime(wbeth_usdt['local_timestamp'])

#Set local_timestamp as index
btc_usdt = btc_usdt.set_index('local_timestamp')
wbtc_usdt = wbtc_usdt.set_index('local_timestamp')
eth_usdt = eth_usdt.set_index('local_timestamp')
wbeth_usdt = wbeth_usdt.set_index('local_timestamp')

```

```python
btc_usdt
```

```python
#Filter date range to 2024-06-01 to 2024-12-31 
btc_usdt = btc_usdt.loc['2024-06-01':'2024-12-31']
wbtc_usdt = wbtc_usdt.loc['2024-06-01':'2024-12-31']
eth_usdt = eth_usdt.loc['2024-06-01':'2024-12-31']
wbeth_usdt = wbeth_usdt.loc['2024-06-01':'2024-12-31']
```

```python
eth_usdt
```

```python
wbeth_usdt
```

```python
#resample data to 1 minute
btc_usdt_1m = btc_usdt.resample('1min').agg({
    'asks[0].price': 'last',
    'asks[0].amount': 'last',
    'bids[0].price': 'last',
    'bids[0].amount': 'last'
})

wbtc_usdt_1m = wbtc_usdt.resample('1min').agg({
    'asks[0].price': 'last',
    'asks[0].amount': 'last',
    'bids[0].price': 'last',
    'bids[0].amount': 'last'
})

eth_usdt_1m = eth_usdt.resample('1min').agg({
    'asks[0].price': 'last',
    'asks[0].amount': 'last',
    'bids[0].price': 'last',
    'bids[0].amount': 'last'
})

wbeth_usdt_1m = wbeth_usdt.resample('1min').agg({
    'asks[0].price': 'last',
    'asks[0].amount': 'last',
    'bids[0].price': 'last',
    'bids[0].amount': 'last'
})

```

```python
eth_usdt_1m
```

```python
btc_usdt_1m.isna().sum()
wbtc_usdt_1m.isna().sum()
eth_usdt_1m.isna().sum()
wbeth_usdt_1m.isna().sum()

print(btc_usdt_1m.isna().sum())
print(wbtc_usdt_1m.isna().sum())
print(eth_usdt_1m.isna().sum())
print(wbeth_usdt_1m.isna().sum())
```

```python
#Drop NaN values
btc_usdt_1m = btc_usdt_1m.dropna()
wbtc_usdt_1m = wbtc_usdt_1m.dropna()
eth_usdt_1m = eth_usdt_1m.dropna()
wbeth_usdt_1m = wbeth_usdt_1m.dropna()
```

```python
#Calculate mid prices
btc_usdt_1m['mid_price'] = (btc_usdt_1m['asks[0].price'] + btc_usdt_1m['bids[0].price']) / 2
wbtc_usdt_1m['mid_price'] = (wbtc_usdt_1m['asks[0].price'] + wbtc_usdt_1m['bids[0].price']) / 2
eth_usdt_1m['mid_price'] = (eth_usdt_1m['asks[0].price'] + eth_usdt_1m['bids[0].price']) / 2
wbeth_usdt_1m['mid_price'] = (wbeth_usdt_1m['asks[0].price'] + wbeth_usdt_1m['bids[0].price']) / 2

# #Calculate liquidity
btc_usdt_1m['ask_liquidity'] = btc_usdt_1m['asks[0].price'] * btc_usdt_1m['asks[0].amount']
btc_usdt_1m['bid_liquidity'] = btc_usdt_1m['bids[0].price'] * btc_usdt_1m['bids[0].amount']

wbtc_usdt_1m['ask_liquidity'] = wbtc_usdt_1m['asks[0].price'] * wbtc_usdt_1m['asks[0].amount']
wbtc_usdt_1m['bid_liquidity'] = wbtc_usdt_1m['bids[0].price'] * wbtc_usdt_1m['bids[0].amount']

eth_usdt_1m['ask_liquidity'] = eth_usdt_1m['asks[0].price'] * eth_usdt_1m['asks[0].amount']
eth_usdt_1m['bid_liquidity'] = eth_usdt_1m['bids[0].price'] * eth_usdt_1m['bids[0].amount']

wbeth_usdt_1m['ask_liquidity'] = wbeth_usdt_1m['asks[0].price'] * wbeth_usdt_1m['asks[0].amount']
wbeth_usdt_1m['bid_liquidity'] = wbeth_usdt_1m['bids[0].price'] * wbeth_usdt_1m['bids[0].amount']

```

```python
btc_usdt_1m
```

```python

#Remove columns before merging

#Bitcoin
#--------------------------------------------------------------
btc_mid_price = btc_usdt_1m['mid_price'].rename('btc_mid_price')
wbtc_mid_price = wbtc_usdt_1m['mid_price'].rename('wbtc_mid_price')

btc_ask_price = btc_usdt_1m['asks[0].price'].rename('btc_ask_price')
btc_bid_price = btc_usdt_1m['bids[0].price'].rename('btc_bid_price')

wbtc_ask_price = wbtc_usdt_1m['asks[0].price'].rename('wbtc_ask_price')
wbtc_bid_price = wbtc_usdt_1m['bids[0].price'].rename('wbtc_bid_price')

btc_ask_liquidity = btc_usdt_1m['ask_liquidity'].rename('btc_ask_liquidity')
btc_bid_liquidity = btc_usdt_1m['bid_liquidity'].rename('btc_bid_liquidity')

wbtc_ask_liquidity = wbtc_usdt_1m['ask_liquidity'].rename('wbtc_ask_liquidity')
wbtc_bid_liquidity = wbtc_usdt_1m['bid_liquidity'].rename('wbtc_bid_liquidity')
#--------------------------------------------------------------


#Ethereum
#--------------------------------------------------------------
eth_mid_price = eth_usdt_1m['mid_price'].rename('eth_mid_price')
wbeth_mid_price = wbeth_usdt_1m['mid_price'].rename('wbeth_mid_price')

eth_ask_price = eth_usdt_1m['asks[0].price'].rename('eth_ask_price')
eth_bid_price = eth_usdt_1m['bids[0].price'].rename('eth_bid_price')

wbeth_ask_price = wbeth_usdt_1m['asks[0].price'].rename('wbeth_ask_price')
wbeth_bid_price = wbeth_usdt_1m['bids[0].price'].rename('wbeth_bid_price')

eth_ask_liquidity = eth_usdt_1m['ask_liquidity'].rename('eth_ask_liquidity')
eth_bid_liquidity = eth_usdt_1m['bid_liquidity'].rename('eth_bid_liquidity')

wbeth_ask_liquidity = wbeth_usdt_1m['ask_liquidity'].rename('wbeth_ask_liquidity')
wbeth_bid_liquidity = wbeth_usdt_1m['bid_liquidity'].rename('wbeth_bid_liquidity')
#--------------------------------------------------------------


```

```python

#Merge dataframes

btc = pd.concat([btc_mid_price, btc_ask_price, btc_bid_price, btc_ask_liquidity, btc_bid_liquidity], axis=1)
wbtc = pd.concat([wbtc_mid_price, wbtc_ask_price, wbtc_bid_price, wbtc_ask_liquidity, wbtc_bid_liquidity], axis=1)

eth = pd.concat([eth_mid_price, eth_ask_price, eth_bid_price, eth_ask_liquidity, eth_bid_liquidity], axis=1)
wbeth = pd.concat([wbeth_mid_price, wbeth_ask_price, wbeth_bid_price, wbeth_ask_liquidity, wbeth_bid_liquidity], axis=1)

#Reset index so that the local_timestamp is a column and not the index for merging
btc = btc.reset_index()
wbtc = wbtc.reset_index()

eth = eth.reset_index()
wbeth = wbeth.reset_index()

btc_wbtc_combined_1m = pd.merge(btc, wbtc, on='local_timestamp', how='inner')
btc_wbtc_combined_1m.set_index('local_timestamp', inplace=True) #restore datetime index

eth_wbeth_combined_1m = pd.merge(eth, wbeth, on='local_timestamp', how='inner')
eth_wbeth_combined_1m.set_index('local_timestamp', inplace=True) #restore datetime index
```
